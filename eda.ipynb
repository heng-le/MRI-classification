```python
# importing relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# look at the dataset 
base_folder = r"mri_scans"
train_folder = os.path.join(base_folder, "Training")
test_folder = os.path.join(base_folder, "Testing")


dirs = os.listdir(train_folder)

dirs = [d for d in dirs if os.path.isdir(os.path.join(train_folder, d))]

dirs.sort()
print(dirs)
```

```python
# plot the first few images in each directory
print("Directories:", dirs)
for d in dirs:
    print("Current Directory:", d)
    files = os.listdir(os.path.join(train_folder, d))
    files = [f for f in files if f.endswith(".jpg") or f.endswith(".jpeg")]
    files.sort()
    print("Files Found:", files)
    for f in files[:3]:
        print("Current File:", f)
        img = plt.imread(os.path.join(train_folder, d, f))
        plt.imshow(img)
        plt.title(d)
        plt.show()
```

```python
# finding out how many pictures in each folder for training and testing
train_counts = {}
test_counts = {}

for d in dirs:
    train_files = os.listdir(os.path.join(train_folder, d))
    train_files = [f for f in train_files if f.endswith(".jpg") or f.endswith(".jpeg")]
    train_counts[d] = len(train_files)
    
    test_files = os.listdir(os.path.join(test_folder, d))
    test_files = [f for f in test_files if f.endswith(".jpg") or f.endswith(".jpeg")]
    test_counts[d] = len(test_files)

print("Train Counts:", train_counts)
print("Test Counts:", test_counts)
```
